```python
import string

# Paragraphs
para1 = """Technology is changing the world rapidly! Every day, new inventions are introduced.
People use smartphones, laptops, and tablets for communication. However, too much technology
can sometimes create problems: stress, distraction, and lack of focus. So, balance is important!"""

para2 = """Students often study late at night; exams can be stressful.
Some students prefer group study, while others like studying alone.
Teachers always say: practice more, revise properly, and stay confident!
Success comes to those who work hard."""

# Combine text
text = para1 + " " + para2

# Display original text
print("Original Text:\n")
print(text)

# Remove punctuation
text_no_punct = text.translate(str.maketrans('', '', string.punctuation))

# Display cleaned text
print("\nText after removing punctuation:\n")
print(text_no_punct)
```

```python
import nltk

# Download stopwords (run once)
nltk.download('stopwords')

# Combine text
text = para1 + " " + para2

# Remove punctuation
text_no_punct = text.translate(str.maketrans('', '', string.punctuation))

# Stop word removal
stop_words = set(stopwords.words('english'))

words = text_no_punct.split()

filtered_words = [word for word in words if word.lower() not in stop_words]

# Display result
print("After removing stop words:\n")
print(filtered_words)
```

```python
from nltk.stem import PorterStemmer

# Initialize stemmer
stemmer = PorterStemmer()

# Apply stemming
stemmed_words = [stemmer.stem(word) for word in filtered_words]

# Display result
print("After Stemming:\n")
print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
import nltk

# Download resources (run once)
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Apply lemmatization
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

# Display result
print("After Lemmatization:\n")
print(lemmatized_words)
```
