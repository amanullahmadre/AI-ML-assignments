```python
from sklearn.feature_extraction.text import CountVectorizer

# Input text
text = ["""Technology is changing the world rapidly! Every day, new inventions are introduced.
People use smartphones, laptops, and tablets for communication. However, too much technology
can sometimes create problems: stress, distraction, and lack of focus. So, balance is important! Students often study late at night; exams can be stressful.
Some students prefer group study, while others like studying alone.
Teachers always say: practice more, revise properly, and stay confident!
Success comes to those who work hard."""]

# Initialize CountVectorizer
bow = CountVectorizer()

# Fit and transform
bow_matrix = bow.fit_transform(text)

# Output
print("Vocabulary:\n", bow.get_feature_names_out())
print("\nBoW Matrix:\n", bow_matrix.toarray())
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(text)

print("Vocabulary:\n", tfidf.get_feature_names_out())
print("\nTF-IDF Matrix:\n", tfidf_matrix.toarray())
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
text2 = """Artificial Intelligence and Machine Learning are rapidly growing technologies in today's world. Many companies are investing heavily in AI research and development. AI helps in automation, data analysis, prediction, and decision making. Machine learning is a subset of AI that focuses on learning from data without being explicitly programmed.

Deep learning is a specialized branch of machine learning that uses neural networks with multiple layers. These models are capable of understanding complex patterns in data such as images, speech, and text. Natural Language Processing allows machines to understand human language and interact with users in a meaningful way.

In industries like healthcare, AI is used for disease prediction and diagnosis. In finance, it helps detect fraud and automate trading. In transportation, self-driving cars are being developed using AI. Education systems are also integrating AI to provide personalized learning experiences for students.

However, there are concerns about job displacement due to automation. Experts believe that AI will create new job opportunities while replacing repetitive tasks. Ethical considerations, data privacy, and bias in AI systems are also important challenges that need to be addressed carefully."""
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python

tokens = word_tokenize(text2.lower())
sentences = [tokens]


model = Word2Vec(sentences, vector_size=100, window=5, min_count=1)


print("ai vs machine:", model.wv.similarity('ai', 'machine'))
print("data vs analysis:", model.wv.similarity('data', 'analysis'))
print("learning vs deep:", model.wv.similarity('learning', 'deep'))
```

```python
print("Similar to ai:\n", model.wv.most_similar('ai'))
print("Similar to data:\n", model.wv.most_similar('data'))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = [
    "Artificial intelligence is powerful",
    "AI is very powerful technology",
    "I love playing cricket"
]

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(sentences)


similarity_matrix = cosine_similarity(vectors)
print(similarity_matrix)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer


texts = [
    "I love this product",
    "This is amazing",
    "I hate this",
    "This is bad",
    "Very good experience",
    "Worst service ever"
]

labels = [1, 1, 0, 0, 1, 0]


vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)


X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2)


model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions)
```

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer


texts = [
    "I love AI",
    "Machine learning is great",
    "I enjoy cricket",
    "Football is fun",
    "Deep learning is powerful",
    "Sports are exciting"
]


vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)


kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
print(kmeans.labels_)
```
