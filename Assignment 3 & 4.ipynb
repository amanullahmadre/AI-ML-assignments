```python

# REQUIRED LIBRARIES
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.svm import SVC
from sklearn.cluster import KMeans

# INITIALIZE EMBEDDING MODEL
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Function: Convert text → vector embedding
def get_embedding(text):
    return embedding_model.encode(text)

# 1. MEASURING SIMILARITY BETWEEN SENTENCES

text_a = get_embedding("NLP is interesting")
text_b = get_embedding("Machine learning is interesting")
text_c = get_embedding("Math is boring")

print("Similarity Scores:\n")

score_ab = cosine_similarity([text_a], [text_b])[0][0]
print("Text A vs Text B:", score_ab)

score_ac = cosine_similarity([text_a], [text_c])[0][0]
print("Text A vs Text C:", score_ac)

score_bc = cosine_similarity([text_b], [text_c])[0][0]
print("Text B vs Text C:", score_bc)

# 2. SENTIMENT CLASSIFICATION USING SVM

data_samples = [
    "NLP is interesting",
    "Machine learning is powerful",
    "I hate studying",
    "Deep learning is boring",
    "Math is boring",
    "Machine learning is advanced",
    "I love studying",
    "Artificial intelligence is boring"
]

# Convert sentences into embeddings
features = np.array([get_embedding(sentence) for sentence in data_samples])

# Target labels (1 = positive, 0 = negative)
targets = [1, 1, 0, 0, 0, 1, 1, 0]

# Train SVM classifier
classifier = SVC()
classifier.fit(features, targets)

# Prediction on new input
new_input = get_embedding("Deep learning is interesting")
prediction = classifier.predict([new_input])

print("\nPredicted Class:", prediction)

# 3. GROUPING SENTENCES USING K-MEANS

clustering_model = KMeans(n_clusters=3, random_state=42)
clustering_model.fit(features)

print("\nCluster Assignments:", clustering_model.labels_)
```

```python

# VEGETABLE PRICE CHAT ASSISTANT

import random

# Vegetable price database (₹ per kg)
price_list = {
    "tomato": 22, "onion": 28, "potato": 18, "carrot": 35, "cabbage": 15,
    "brinjal": 20, "beans": 30, "capsicum": 40, "cauliflower": 25, "peas": 45,
    "spinach": 10, "radish": 12, "beetroot": 18, "cucumber": 20, "pumpkin": 16,
    "bottle gourd": 14, "ridge gourd": 18, "ladyfinger": 26, "chilli": 50,
    "ginger": 60, "garlic": 80
}

# Greeting responses
def random_greeting():
    options = ["Hello!", "Hi!", "Hey there!"]
    return random.choice(options)

# Identify vegetable from input text
def extract_veg(user_text):
    user_text = user_text.lower()
    for name in price_list:
        if name in user_text:
            return name
    return None

# Generate price reply
def build_price_reply(veg_name):
    amount = price_list[veg_name]
    return f"{veg_name.title()} is available at ₹{amount} per kg."

# Chat loop
def run_bot():
    print("Bot: I can provide vegetable prices. Ask me anything!")

    while True:
        query = input("You: ").strip().lower()

        # Exit condition
        if query in ["exit", "bye", "quit"]:
            print("Bot: Goodbye! Take care.")
            break

        # Greeting check
        if any(word in query for word in ["hi", "hello", "hey"]):
            print("Bot:", random_greeting())
            continue

        # Price-related keywords
        if any(word in query for word in ["price", "cost", "rate"]):
            veg_item = extract_veg(query)
            if veg_item:
                print("Bot:", build_price_reply(veg_item))
            else:
                print("Bot: Please tell me which vegetable you're asking about.")
            continue

        # If only vegetable name is mentioned
        veg_item = extract_veg(query)
        if veg_item:
            print("Bot:", build_price_reply(veg_item))
        else:
            print("Bot: I can help with prices of vegetables like tomato, onion, potato, etc.")

# Start chatbot
run_bot()
```
